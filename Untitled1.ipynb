```python
import icclim
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs 
import numpy as np

summer_days=icclim.su("data/latest/tasmax_day_CanESM5_ssp585_r1i1p1f1_gn_21010101-21801231.nc", out_file="summer_days.nc")
```

```python
data = xr.open_dataset('summer_days.nc')
```

```python
time_avg = data.mean(dim='time')
```

```python
fig = plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()

time_avg.to_array().plot(ax=ax, transform=ccrs.PlateCarree())

plt.title('Time Average of Summer days')
plt.show()
```
